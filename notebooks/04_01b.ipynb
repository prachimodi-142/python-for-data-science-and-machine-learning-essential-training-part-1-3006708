```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

```python
data = {'names' : ['prachi', 'arvind', 'akshay', 'shahrukh', 'vivek', 'sahu', 'supriya', 'divya'],
        'age' : [26,30,31,29,28,32,31,28],
        'gender' : ['female', 'male', 'male', 'male', 'male', 'male', 'female', 'female'],
        'rank' : [1,3,2,8,5,6,4,7]}

df = DataFrame(data)
df
```

### Matplotlib's Bar Chart

```python
plt.bar(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing age')
plt.show()
```

```python
plot = sns.barplot(data = df, x = 'names', y = 'age')
plot.set_title("Comparing Ages")
plt.show()
```

### Line Plot Matplotlib

```python
plt.plot(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing age')
plt.show()
```

### Line Plot Seaborn

```python
plot = sns.lineplot(data = df, x = 'names', y = 'age')
plot.set_title("Comparing Ages")
plt.show()
```

### Pie Chart Matplotlib

```python
plt.pie(df['age'], labels = df ['names'])
plt.title('Comparing age')
plt.show()
```

### Pie Chart Seaborn

```python
colors = sns.color_palette('pastel')[0:5]
plt.pie(df['age'], labels = df['names'], colors=colors)
plt.show()
```
